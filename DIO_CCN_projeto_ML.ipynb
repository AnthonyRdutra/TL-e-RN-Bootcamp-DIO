```python
%matplotlib inline

import os

import random
import numpy as np
import keras

import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow

from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from keras.models import Model


root = '/content/kagglecatsanddogs_5340/'
train_split, val_split = 0.7, 0.15

categories = [os.path.join(root, 'PetImages', 'Cat'),
              os.path.join(root, 'PetImages', 'Dog')]


def get_image(path):
    try:
        img = image.load_img(path, target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        return img, x
    except Exception as e:
        print(f"Error loading image {path}: {e}")
        return None, None

data = []
for c, category in enumerate(categories):
    images = [os.path.join(dp, f) for dp, dn, filenames
              in os.walk(category) for f in filenames
              if os.path.splitext(f)[1].lower() in ['.jpg','.png','.jpeg']]


    for img_path in images:
        img, x = get_image(img_path)
        if img is not None:
            data.append({'x':np.array(x[0]), 'y':c})

# # count the number of classes
num_classes = len(categories)

random.shuffle(data)

idx_val = int(train_split * len(data))
idx_test = int((train_split + val_split) * len(data))
train = data[:idx_val]
val = data[idx_val:idx_test]

test = data[idx_test:]

x_train, y_train = np.array([t["x"] for t in train]), [t["y"] for t in train]
x_val, y_val = np.array([t["x"] for t in val]), [t["y"] for t in val]
x_test, y_test = np.array([t["x"] for t in test]), [t["y"] for t in test]
# print(y_test)


# normalize data
x_train = x_train.astype('float32') / 255.
x_val = x_val.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# convert labels to one-hot vectors
y_train = keras.utils.to_categorical(y_train, num_classes)
y_val = keras.utils.to_categorical(y_val, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print(y_test.shape)


# summary
print("finished loading %d images from %d categories"%(len(data), num_classes))
print("train / validation / test split: %d, %d, %d"%(len(x_train), len(x_val), len(x_test)))
print("training data shape: ", x_train.shape)
print("training labels shape: ", y_train.shape)
print("process complete")


images = [os.path.join(dp, f) for dp, dn, filenames in os.walk(root) for f in filenames if os.path.splitext(f)[1].lower() in ['.jpg','.png','.jpeg']]
idx = [int(len(images) * random.random()) for i in range(8)]
imgs = [image.load_img(images[i], target_size=(224, 224)) for i in idx]
concat_image = np.concatenate([np.asarray(img) for img in imgs], axis=1)
plt.figure(figsize=(16,4))
plt.imshow(concat_image)




# build the network
model = Sequential()
print("Input dimensions: ",x_train.shape[1:])

model.add(Conv2D(32, (3, 3), input_shape=x_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.summary()

# # compile the model to use categorical cross-entropy loss function and adadelta optimizer
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=128,
                    epochs=10,
                    validation_data=(x_val, y_val))


fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(121)
ax.plot(history.history["val_loss"])
ax.set_title("validation loss")
ax.set_xlabel("epochs")

ax2 = fig.add_subplot(122)
ax2.plot(history.history["val_accuracy"])
ax2.set_title("validation accuracy")
ax2.set_xlabel("epochs")
ax2.set_ylim(0, 1)

plt.show()


loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy)

vgg = keras.applications.VGG16(weights='imagenet', include_top=True)
vgg.summary()


# make a reference to VGG's input layer
inp = vgg.input

# make a new softmax layer with num_classes neurons
new_classification_layer = Dense(num_classes, activation='softmax')

# connect our new layer to the second to last layer in VGG, and make a reference to it
out = new_classification_layer(vgg.layers[-2].output)

# create a new network between inp and out
model_new = Model(inp, out)

# make all layers untrainable by freezing weights (except for last layer)
for l, layer in enumerate(model_new.layers[:-1]):
    layer.trainable = False

# ensure the last layer is trainable/not frozen
for l, layer in enumerate(model_new.layers[-1:]):
    layer.trainable = True

model_new.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model_new.summary()

history2 = model_new.fit(x_train, y_train,
                         batch_size=128,
                         epochs=10,
                         validation_data=(x_val, y_val))

fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(121)
ax.plot(history.history["val_loss"])
ax.plot(history2.history["val_loss"])
ax.set_title("validation loss")
ax.set_xlabel("epochs")

ax2 = fig.add_subplot(122)
ax2.plot(history.history["val_accuracy"])
ax2.plot(history2.history["val_accuracy"])
ax2.set_title("validation accuracy")
ax2.set_xlabel("epochs")
ax2.set_ylim(0, 1)

plt.show()

loss, accuracy = model_new.evaluate(x_test, y_test, verbose=0)

print('Test loss:', loss)
print('Test accuracy:', accuracy)


#Teste do treinamento com 3 samples
samples = ['/content/sample_data/Design-sem-nome-32-2.png', '/content/sample_data/Gato-tigrado-6.jpeg', '/content/sample_data/sitesdefaultfilesstylessquare_medium_440x440public2022-07Bullmastiff.jpg']


for path in samples:
    img, x = get_image(path)
    if x is not None:
        probabilities = model_new.predict(x)
        print(f"Predições para {path}:\n", probabilities)
    else:
        print(f"Falha ao carregar: {path}")


# img, x = get_image('content/sample-data/Gato-tigrado-6.jpeg')
# probabilities = model_new.predict([x])
```

Extracts zip into a folder and reduces the dataset os it doenst get cap-out the ram

```python
import zipfile
import os
import random

def clean_data_set(pasta):
    imagens_validas = []

    # Coleta apenas imagens com nome numérico
    for nome_arquivo in os.listdir(pasta):
        nome, ext = os.path.splitext(nome_arquivo)
        if ext.lower() in ['.jpg', '.jpeg', '.png']:
            try:
                int(nome)  # Garante que o nome é numérico
                imagens_validas.append(nome_arquivo)
            except ValueError:
                continue

    total = len(imagens_validas)

    if total <= 200:
        print(f"Apenas {total} imagens encontradas em {pasta}. Nenhuma imagem removida.")
        return

    # Embaralha e seleciona imagens para manter
    random.shuffle(imagens_validas)
    imagens_para_manter = set(imagens_validas[:200])

    # Remove o restante
    removidas = 0
    for nome_arquivo in imagens_validas:
        if nome_arquivo not in imagens_para_manter:
            caminho_completo = os.path.join(pasta, nome_arquivo)
            try:
                os.remove(caminho_completo)
                removidas += 1
            except Exception as e:
                print(f"Erro ao remover {nome_arquivo}: {e}")

    print(f"Removidas {removidas} imagens da pasta: {pasta}")


try:
  # Caminho para o arquivo .zip
  zip_path = 'kagglecatsanddogs_5340.zip'
  extract_to = 'kagglecatsanddogs_5340'
  os.makedirs(extract_to, exist_ok=True)

  with zipfile.ZipFile(zip_path, 'r') as zip_ref:
      zip_ref.extractall(extract_to)

  print("Extração concluída.")


  # Caminhos das pastas de gatos e cachorros
  paths = ['/content/kagglecatsanddogs_5340/PetImages/Dog', '/content/kagglecatsanddogs_5340/PetImages/Dog']
  i = 0
  while i < len(paths):
    clean_data_set(paths[i])
    i += 1

except Exception as e:
  print(e)
```
